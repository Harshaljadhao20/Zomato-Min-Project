```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Zomato data .csv")
```

```python
df
```

```python
df.info()
```

Convert the data type of column - rate

```python
def handleRate(value) :
  value = str(value).split("/")
  value = value[0]
  return float(value)

df["rate"] = df["rate"].apply(handleRate)
print(df.head())
```

Q1) What type of restsurant do the majoruty of customers order from ?

```python
df.head(5)
```

```python
sns.countplot(x=df["listed_in(type)"])
plt.xlabel("Type of restsurant")
```

Conclusion :- majority of the resturant falls in dinning category

Q2) How many votes has each type of restaurant received from customers ?

```python
df.head(5)
```

```python
grouped_df = df.groupby("listed_in(type)")["votes"].sum()
result = pd.DataFrame({"votes" : grouped_df})
plt.plot(result, c= "red", marker = "o")
plt.xlabel("Type of restaurant ", c="b", size=20)
plt.ylabel("votes", c="green", size=20)
```

Conclusion :- dinning resturants has recieved maximum vates

Q3) What are the ratings that manjority of restaurants  have received ?

Q5) Which mode (online or offline) has received the maximum rating ?

 Q6) Which type of restaurant recived more offline orders, os that Zomato cam=n provide customers with some good offers ?

```python
pivot_table = df.pivot_table(index="listed_in(type)", columns="online_order", aggfunc="size", fill_value=0)

sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt="d")
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (type)")
plt.show()
```

Conclusioin :- Dining restarants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clints preferns orders in person at restaurantsn, but prefe online ordering at cafes.
